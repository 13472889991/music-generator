```
denver = pd.read_csv("/content/drive/My Drive/Data Dump/Denver-Consumer.csv")
miami = pd.read_csv("/content/drive/My Drive/Data Dump/Miami-Consumer.data.csv")
portland = pd.read_csv("/content/drive/My Drive/Data Dump/Portland Consumer Data.csv")
drider = pd.read_csv('/content/drive/My Drive/Data Dump/Denver Rider Data.csv')
mrider = pd.read_csv('/content/drive/My Drive/Data Dump/Miami Rider Data.csv')
prider = pd.read_csv('/content/drive/My Drive/Data Dump/Portland Rider Data.csv')
total = pd.concat([drider, mrider,prider], ignore_index=True)
```

```
# We can merge the dataframe of rider and consumer by first name and last name
def merge_df(d1,d2):
  d1['first_name'] = d1['Rider First Name']
  mergedStuff = pd.merge(d2, d1, on=['first_name'], how='inner')
  mergedStuff['lastname2'] = mergedStuff['Rider Last Name']
  return mergedStuff.query('last_name == lastname2')
def recast_salary(df1):
  df1['Annual Salary'] = df1['Annual Salary'].str.replace(',', '').str.replace('$', '').astype(float)
  return df1
def recast_debt(df1):
  df1['Total Debt'] = df1['Total Debt'].str.replace(',', '').str.replace('$', '').astype(float)
  return df1
def recast_trip(df1):
    df1['Trip Cost'] = df1['Trip Cost'].str.replace(',', '').str.replace('$', '').astype(float)
    return df1

```

```
drider = recast_trip(drider)
mrider = recast_trip(mrider)
prider= recast_trip(prider)
denver = recast_salary(recast_debt(denver))
portland = recast_salary(recast_debt(portland))
miami = recast_salary(recast_debt(miami))
```

```
x=drider['Trip Cost'].sum()/drider['Trip Distance (in Miles)'].sum()
y=mrider['Trip Cost'].sum()/mrider['Trip Distance (in Miles)'].sum()
z = prider['Trip Cost'].sum()/prider['Trip Distance (in Miles)'].sum()
```

```
def analyze_riders(df1):
  taxi = df1[df1['Vendor'] == 'Taxi']
  uber = df1[df1['Vendor'] == 'Uber']
  lyft = df1[df1['Vendor'] == 'Lyft']
  print('Taxi,Uber,Lyft Comparisons')
  print(df1['Vendor'].value_counts())
  print('Mean cost:  ' + str(taxi['Trip Cost'].mean()) +","+str(uber['Trip Cost'].mean()) +"," + str(lyft['Trip Cost'].mean()))
  print('Mean Distance:   '+ str(taxi['Trip Distance (in Miles)'].mean()) +","+str(uber['Trip Distance (in Miles)'].mean()) +"," + str(lyft['Trip Distance (in Miles)'].mean()))
  print('Mean cost/distance:   '+ str(taxi['Trip Cost'].sum()/taxi['Trip Distance (in Miles)'].sum()) +","+str(uber['Trip Cost'].sum()/uber['Trip Distance (in Miles)'].sum()) +"," + str(lyft['Trip Cost'].sum()/lyft['Trip Distance (in Miles)'].sum()))
```

```
analyze_riders(total)
```

```
analyze_riders(mrider)
```
